```python
%matplotlib inline
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn import preprocessing
from sklearn.preprocessing import RobustScaler

# Import Tensorflow libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import MaxPooling2D

from tensorflow.keras.utils import to_categorical


###### Tensorflow-GPU ########
physical_devices = tf.config.experimental.list_physical_devices('GPU')
if physical_devices:  # Check if the list is not empty
    tf.config.experimental.set_memory_growth(physical_devices[0], True)
else:
    print("No GPU devices found.")

import re
```

```python
# Import dataset
dataset = pd.read_csv('cancer.csv')
dataset.head(20)
```

```python
# Preprocessing 
dataset.drop('id',axis=1, inplace=True)

# Drop the unnamed column
dataset.drop(dataset.columns[31],axis=1,inplace=True)

# diagnosis labels M -> 1 and B -> 0
dataset["diagnosis"] = dataset.diagnosis.replace(['M', 'B'], [1, 0])
```

```python
# Split into features and targets

y = dataset.diagnosis
X = dataset.drop('diagnosis',axis=1)
```

```python
# Create training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state =0)

# convert to numpy array
X_train = X_train.values
X_test = X_test.values

# One-hot encoding of the targets
#y_train = to_categorical(y_train)

# Normalize the features
# via robustscaler: This Scaler removes the median and scales the data according to the quantile range 

Rscaler = RobustScaler()
Rscaler.fit(X_train)

X_train = Rscaler.transform(X_train)
X_test = Rscaler. transform(X_test)
```

```python
# Import necessary libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Activation, Dropout
from tensorflow.keras.optimizers import Adam, SGD
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Parameters for the neural network
dropoutrate = 0.2  # Dropout rate to prevent overfitting
batchsize = 16     # Number of samples per batch during training
inputdim = X_train.shape[1]  # Number of input features (30 in this case)

# Optimizers
adam = tf.keras.optimizers.Adam()  # Adam optimizer for adaptive learning rate
sgd = tf.keras.optimizers.SGD()    # SGD optimizer (not used in this code)

# Initialize the Sequential neural network
model = Sequential()

# Input layer with 100 neurons, uniform weight initialization
model.add(Dense(100, input_dim=inputdim, kernel_initializer='uniform'))
model.add(BatchNormalization())  # Batch normalization to stabilize training
model.add(Activation('relu'))    # ReLU activation function
model.add(Dropout(dropoutrate))  # Dropout to prevent overfitting

# First hidden layer with 100 neurons
model.add(BatchNormalization())  # Batch normalization
model.add(Dense(100, kernel_initializer='uniform'))  # Dense layer
model.add(BatchNormalization())  # Batch normalization
model.add(Activation('relu'))    # ReLU activation function
model.add(Dropout(dropoutrate))  # Dropout to prevent overfitting

# Second hidden layer with 100 neurons
model.add(Dense(100, kernel_initializer='uniform'))  # Dense layer
model.add(BatchNormalization())  # Batch normalization
model.add(Activation('relu'))    # ReLU activation function
model.add(Dropout(dropoutrate))  # Dropout to prevent overfitting

# Output layer with 1 neuron and sigmoid activation for binary classification
model.add(Dense(1, activation='sigmoid'))

# Compile the model
# Loss function: Binary cross-entropy for binary classification
# Optimizer: Adam optimizer
# Metrics: Accuracy to evaluate model performance
model.compile(loss=tf.keras.losses.binary_crossentropy, optimizer=adam, metrics=['accuracy'])

# Train the model
# Training data: X_train and y_train
# Validation split: 30% of the training data is used for validation
# Epochs: Train for 80 epochs
# Batch size: 16 samples per batch
# Verbose: 1 (displays training progress)
history = model.fit(X_train, y_train, epochs=80, validation_split=0.3, batch_size=batchsize, verbose=1)

# Testing the model
# Predict probabilities for the test data
# Convert probabilities to binary predictions (1 if > 0.5, else 0)
y_pred = (model.predict(X_test) > 0.5).astype('int32')

# Print classification report
print('\n')
print(classification_report(y_test, y_pred))

# Compute and print the confusion matrix
cf = confusion_matrix(y_test, y_pred)
print(cf)

# Compute and print the accuracy score
print(accuracy_score(y_test, y_pred) * 100)
```

```python
# Plot history

# Accuray 
plt.plot(history.history['accuracy'],'r')
plt.plot(history.history['val_accuracy'],'b')
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# Loss 
plt.plot(history.history['loss'],'r')
plt.plot(history.history['val_loss'],'b')

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
